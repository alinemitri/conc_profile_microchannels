```python
from skimage import exposure
import numpy as np
import matplotlib.pyplot as plt

# Example image (a simple grayscale gradient)
image = np.linspace(0, 1, 256).reshape(16, 16)

# Rescale intensity: equivalent to MATLAB's imadjust
# Input range: (0.2, 0.8), Output range: (0, 1)
adjusted_image = exposure.rescale_intensity(image, in_range=(0.2, 0.8), out_range=(0, 1))

# Display original and adjusted images
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original Image")
ax[1].imshow(adjusted_image, cmap='gray')
ax[1].set_title("Adjusted Image")
plt.show()
```

```python
from skimage import io, exposure
import matplotlib.pyplot as plt

# Load the JPG image
image = io.imread('Teste_1.jpg')  # Replace with your file path

# Check if the image is in color (RGB), convert to grayscale if needed
if image.ndim == 3:  # Check if it's a color image
    from skimage.color import rgb2gray
    image_gray = rgb2gray(image)
else:
    image_gray = image  # It's already grayscale

# Perform intensity adjustment
# For example, stretch intensity from (0.2, 0.8) to (0, 1)
adjusted_image = exposure.rescale_intensity(image_gray, in_range=(0.2, 0.8), out_range=(0, 1))

# Display the images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_gray, cmap='gray')
ax[0].set_title('Original Grayscale Image - Case 1a')
ax[1].imshow(adjusted_image, cmap='gray')
ax[1].set_title('Adjusted Image')
plt.show()
```

```python
import numpy as np
from scipy.ndimage import map_coordinates
from skimage import io
import matplotlib.pyplot as plt

# Load the image
image = io.imread('Teste_1.jpg', as_gray=True)  # Grayscale

# Define the start and end points of the line
start = (230, 300)  # (row, col)
end = (320, 300)

# Generate x and y coordinates along the line
num_points = 100
x = np.linspace(start[0], end[0], num_points)
y = np.linspace(start[1], end[1], num_points)

# Sample the intensity values along the line
line_profile = map_coordinates(image, [x, y], order=1)

# Plot the line profile
plt.figure(figsize=(8, 4))
plt.plot(line_profile, label='Intensity Profile')
plt.xlabel('Distance along line')
plt.ylabel('Intensity')
plt.title('Line Intensity Profile - Case 1a')
plt.legend()
plt.show()
```
